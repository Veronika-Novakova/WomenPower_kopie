```python
import pandas as pd

# Formatovani data
breclav_original = pd.read_csv("open-meteo-Breclav_clean.csv")
breclav_original["time"] = pd.to_datetime(breclav_original["time"])
# Testovaci vystup
print(breclav_original.head())
print(breclav_original.shape)
```

```python
# Zde dosadit datum pranostiky
datum_pranostiky = "02-07"

filtrovane_dny_pranostiky = []
# 50 let ve statistice, od roku 1975
for i in range(50):
    current_year = i+1975
    # Vybrat pouze hodiny v datech pranostiky, pripojit po jednotlivych rocich do filtrovane_dny_pranostiky
    filtrovane_dny_pranostiky.append( breclav_original[breclav_original["time"].dt.date == pd.to_datetime(str(current_year)+"-"+datum_pranostiky).date()] )
print("filtrovane_dny_pranostiky len: ", len(filtrovane_dny_pranostiky))
```

```python
final_vykazy_dnu = []
for i in range(50):
    # Vynulovani promennych pro soucet hodnot v jednom dni v roce
    teplota = 0
    dest = 0
    snih = 0
    # Soucet teplot, srazek a snehu
    for line in filtrovane_dny_pranostiky[i].iloc[0:].iterrows():
        aktualni_rok = line[1].iloc[0].date().year
        teplota += float(line[1].iloc[1])
        dest += float(line[1].iloc[2])
        snih += float(line[1].iloc[3])
    # potrebujeme prumer teploty za 24h 
    teplota /= 24

    pranostika_true = False
    # Zde nastavit podminku pro platnost pranostiky
    if (teplota > 0):
        pranostika_true = True

    # Poskladani jednoho vykazu dne a pripojeni do finalniho vykazu
    vykaz_dne = {"rok": aktualni_rok, "teplota": teplota, "dest": dest, "snih": snih, "true": pranostika_true}
    final_vykazy_dnu.append(vykaz_dne)
print("final_vykazy_dnu: ", final_vykazy_dnu)
print("final_vykazy_dnu len: ", len(final_vykazy_dnu))

# Exportovat final_vykaz_dnu do CSV pro praci v PowerBI
import csv

# Vybrat nazev output .csv souboru dle lokace a data pranostiky
with open("output.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=final_vykazy_dnu[0].keys())
    writer.writeheader()      # write column names
    writer.writerows(final_vykazy_dnu)    # write all rows
```
