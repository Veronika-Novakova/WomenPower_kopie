```python
import pandas as pd

# Formatovani data
breclav_original = pd.read_csv("open-meteo-Breclav_clean.csv")
breclav_original["time"] = pd.to_datetime(breclav_original["time"])
# Testovaci vystup
print(breclav_original.head())
print(breclav_original.shape)
```

```python
# Zde dosadit datum pranostiky ve formatu "MM-DD"
datum_pranostiky = "11-25"

filtrovane_dny_pranostiky_11_25 = []
# 50 let ve statistice, od roku 1975
for i in range(50):
    current_year = i+1975
    # Vybrat pouze hodiny v datech pranostiky, pripojit po jednotlivych rocich do filtrovane_dny_pranostiky_11_25
    filtrovane_dny_pranostiky_11_25.append( breclav_original[breclav_original["time"].dt.date == pd.to_datetime(str(current_year)+"-"+datum_pranostiky).date()] )
# testovaci vystup
print("filtrovane_dny_pranostiky_11_25 len: ", len(filtrovane_dny_pranostiky_11_25))

datum_pranostiky = "12-24"
filtrovane_dny_pranostiky_12_24 = []
# 50 let ve statistice, od roku 1975
for i in range(50):
    current_year = i+1975
    # Vybrat pouze hodiny v datech pranostiky, pripojit po jednotlivych rocich do filtrovane_dny_pranostiky_12_24
    filtrovane_dny_pranostiky_12_24.append( breclav_original[breclav_original["time"].dt.date == pd.to_datetime(str(current_year)+"-"+datum_pranostiky).date()] )
# testovaci vystup
print("filtrovane_dny_pranostiky_12_24 len: ", len(filtrovane_dny_pranostiky_12_24))
```

```python
final_vykazy_dnu_11_25 = []

# testovaci promenna poctu true
true_dny = 0

for i in range(50):
    # Vynulovani promennych pro soucet hodnot v jednom dni v roce
    teplota = 0
    dest = 0
    snih = 0
    # Soucet teplot, srazek a snehu dne 11_25
    for line in filtrovane_dny_pranostiky_11_25[i].iloc[0:].iterrows():
        aktualni_rok = line[1].iloc[0].date().year
        teplota += float(line[1].iloc[1])
        dest += float(line[1].iloc[2])
        snih += float(line[1].iloc[3])
    # potrebujeme prumer teploty za 24h 
    teplota /= 24

    pranostika_true = False
    # Zde nastavit podminku pro platnost pranostiky
    if (teplota > 0 and dest > 0.1):
        pranostika_true = True
        true_dny += 1

    # Poskladani jednoho vykazu dne a pripojeni do finalniho vykazu
    vykaz_dne = {"rok": aktualni_rok, "teplota_11_25": teplota, "dest_11_25": dest, "snih_11_25": snih, "true": pranostika_true}
    final_vykazy_dnu_11_25.append(vykaz_dne)

# Testovaci vypisy pro 11_25
print("true dny: ", true_dny)
print("final_vykazy_dnu_11_25: ", final_vykazy_dnu_11_25)
print("final_vykazy_dnu_11_25 len: ", len(final_vykazy_dnu_11_25))

# To same pro datum 12-24
final_vykazy_obou_dnu = []

# testovaci promenna poctu true
true_dny = 0

for i in range(50):
    # Vynulovani promennych pro soucet hodnot v jednom dni v roce
    teplota = 0
    dest = 0
    snih = 0
    # Soucet teplot, srazek a snehu
    for line in filtrovane_dny_pranostiky_12_24[i].iloc[0:].iterrows():
        aktualni_rok = line[1].iloc[0].date().year
        teplota += float(line[1].iloc[1])
        dest += float(line[1].iloc[2])
        snih += float(line[1].iloc[3])
    # potrebujeme prumer teploty za 24h 
    teplota /= 24

    pranostika_true = False
    # Zde nastavit podminku pro platnost pranostiky
    if (final_vykazy_dnu_11_25[i]['teplota_11_25'] > 0 and final_vykazy_dnu_11_25[i]['dest_11_25'] > 0.1 and teplota < 0):
        pranostika_true = True
        true_dny += 1

    # Poskladani jednoho vykazu dne a vykazu z druheho dne pranostiky a pripojeni do finalniho vykazu obou dnu
    vykaz_dne = {"rok": aktualni_rok, "teplota_11_25": final_vykazy_dnu_11_25[i]['teplota_11_25'], "dest_11_25": final_vykazy_dnu_11_25[i]['dest_11_25'], "snih_11_25": final_vykazy_dnu_11_25[i]['snih_11_25'], "teplota_12_24": teplota, "dest_12_24": dest, "snih_12_24": snih,"true": pranostika_true}
    final_vykazy_obou_dnu.append(vykaz_dne)

# Testovaci vypisy 12_24
print("true dny oba: ", true_dny)
print("final_vykazy_obou_dnu: ", final_vykazy_obou_dnu)
print("final_vykazy_obou_dnu len: ", len(final_vykazy_obou_dnu))

# Exportovat final_vykaz_dnu do CSV pro praci v PowerBI
 
```
